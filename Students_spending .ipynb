```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_stud_spending = pd.read_csv('student_spending.csv')
```

```python
df_stud_spending
```

```python
#remove the unnamed column
del df_stud_spending['Unnamed: 0']
```

```python
df_stud_spending.info()
```

```python
# Create a new dataframe focused on only the columns to be used for this descriptive analysis
# Retain on ['monthly_income','preferred_payment_method','age','personal_care','year_in_school','technology', 'gender']
df_new = df_stud_spending[['monthly_income','preferred_payment_method','age','personal_care','year_in_school','technology', 'gender']].copy()
```

```python
#Analyse data using Correlation
corr_matrix = df_new[['age','gender','year_in_school','personal_care','technology','preferred_payment_method','monthly_income']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Visualise distribution of preferred payment method
sns.set(style="whitegrid")
plt.figure(figsize=(6, 6))
ax = sns.countplot(x='preferred_payment_method', data=df_new, palette='muted')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')


plt.title('Distribution of Preferred Payment Method')
plt.show()
```

```python
sns.set(style="darkgrid")

# Create a pie chart for 'technology'
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(df_new['age'].value_counts(), labels=df_new['age'].value_counts().index, autopct='%1.2f%%', colors=sns.color_palette('pastel'), startangle=90)

# Set plot title
ax.set_title('Distribution of Students Age')

# Display the plot
plt.show()
```

```python
#Using Kdeplot to provide insights into the distribution of students income.
sns.kdeplot(df_new['monthly_income'], shade=True)
plt.title('Density plot of students monthly_income')
plt.xlabel('monthly_income')
plt.ylabel('Density')
plt.show()
```

```python
sns.set(style="darkgrid")

# Create a pie chart for 'technology'
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(df_new['gender'].value_counts(), labels=df_new['gender'].value_counts().index, autopct='%1.2f%%', colors=sns.color_palette('pastel'), startangle=90)

# Set plot title
ax.set_title('Distribution of Students by Gender')

# Display the plot
plt.show()
```

```python
sns.barplot(x=df_new['year_in_school'], y=df_new['technology'])
plt.xlabel('year_in_school')
plt.ylabel('technology')
plt.show()
```

```python
sns.kdeplot(df_new['personal_care'], shade=True)
plt.title('Density plot of students spending on personal_care')
plt.xlabel('personal_care')
plt.ylabel('Density')
plt.show()
```
